```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
size = 9
class_counts = df['Class'].value_counts()

# Построим столбчатую диаграмму
class_counts.plot(kind='bar', figsize=(6, 4))
plt.title('Количество наблюдений для каждого значения Class', size=size)
plt.xlabel('Class', size=size)
plt.ylabel('Количество наблюдений', size=size)
plt.show()
```

```python
# Построим столбчатую диаграмму с логарифмическим масштабом
size = 9
class_counts.plot(kind='bar', log=True, figsize=(6, 4))
plt.title('Количество наблюдений для каждого значения Class (логарифмический масштаб)', size=size)
plt.xlabel('Class', size=size)
plt.ylabel('Количество наблюдений (логарифмическая шкала)',size=size)
plt.show()
```

```python
# Построим две гистограммы по значениям признака V1
legend_font = {
    "size": 10,
    "family": "serif",
}
size = 12
plt.hist(df[df['Class'] == 0]['V1'], bins=20, density=True, alpha=0.5, color='grey', label='Class 0')
plt.hist(df[df['Class'] == 1]['V1'], bins=20, density=True, alpha=0.5, color='red', label='Class 1')
plt.xlabel('V1', size=size)
plt.ylabel('Плотность распределения', size=size)
plt.legend(frameon=False, prop=legend_font)
plt.title('Гистограммы по значениям признака V1', size=size)

plt.show()
```

__Задание № 4__

**Можно ли массив Numpy, состоящий из одного столбца и 12 строк, назвать одномерным?**

Фактически, в NumPy массив, состоящий из одного столбца и нескольких строк, обычно называется двумерным массивом. Это потому, что он представляет собой структуру данных с двумя измерениями: строки и столбцы.

Однако, в некоторых контекстах такой массив может быть рассмотрен как одномерный, если он используется для хранения последовательности данных без внутренней структуры. Например, если массив используется для хранения временного ряда или значения функции в различных точках, то его можно рассматривать как одномерный массив.
